```python
from matplotlib import pyplot as plt
from IPython import display
import time

%matplotlib inline
for i in range(5):
    # any figure
    fig = plt.figure()
    ax = fig.add_subplot(111)
    ax.plot(range(i), range(i))
    display.display(fig)

    # do something that takes a long time:
    time.sleep(3)
```
